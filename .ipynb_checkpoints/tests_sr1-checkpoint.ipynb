```python
dset = st.select_runs(include_tags=['_sr1_preliminary'],run_mode='tpc_kr83m',
                      available=['distinct_channels','event_info'],
                     )
```

```python
columns =['time', 'endtime','drift_time', 's1_a_area', 's2_a_x_cnn', 's2_a_y_cnn', 's2_a_x_gcn', 's2_a_y_gcn', 's2_a_x_mlp', 's2_a_y_mlp', 'z_naive', 'z_dv_corr', 'z', 'theta', 'y', 'x', 'r']
df = st.get_df(dset['name'], targets=('event_info_double','cuts_minimal','cut_Kr_SingleS1S2'), selection_str=['cuts_minimal','cut_Kr_SingleS1S2'], keep_columns=columns)
```

```python
# Bin without a single event! -146.7>z>-147.7,6.0>r>0.0, -1.0>theta>-3.1
conta = 0
for z,r,theta in zip(df['z'],df['r'],df['theta']):
    if (z>-147.6515) & (z<-146.6515) & (r>0.0) & (r<6.0) & (theta > -3.1) & (theta < -1.0):
        conta += 1
print('eventi nel bin: ', conta, ' su ', len(df['z']))

conta = 0
for z in df['z']:
    if (z>-147.6515) & (z<-146.6515):
        conta += 1
print('eventi nel bin: ', conta, ' su ', len(df['z']))

conta = 0
for r in df['r']:
    if (r>0.0) & (r<6.0):
        conta += 1
print('eventi nel bin: ', conta, ' su ', len(df['r']))

conta = 0
for theta in df['theta']:
    if (theta>-3.1) & (theta<-1.0):
        conta += 1
print('eventi nel bin: ', conta, ' su ', len(df['theta']))
```

```python
# Create 2D plot
plt.scatter(df['r'], df['z'], c='r', marker='.')

# Add horizontal dotted lines at specific values of 'r'
r_values = [0.0, 6.04, 12.08, 18.12, 24.16, 30.2, 36.24, 42.28, 48.32, 54.36, 60.4, 66.4]
for r_val in r_values:
    plt.axvline(x=r_val, linestyle='--', color='green')

# Add vertical dotted lines at specific values of 'z'
z_values = [-147.6515, -146.6515, -145.6515, -127.6950625, -109.738625, 
            -91.7821875, -73.82575, -55.8693125, -37.912875, -19.9564375, 
            -2, -1, 0]
for z_val in z_values:
    plt.axhline(y=z_val, linestyle='--', color='green')

# Set labels and title
plt.xlabel('r')
plt.ylabel('z')
plt.title('2D Plot of Events')

plt.show()
```

```python
# Filter DataFrame
filtered_df = df[(df['z'] > -147.6515) & (df['z'] < -146.6515)]
```

```python
# Create 2D plot
plt.scatter(filtered_df['r'], filtered_df['z'], c='r', marker='.')

# Add horizontal dotted lines at specific values of 'r'
r_values = [0.0, 6.04, 12.08, 18.12, 24.16, 30.2, 36.24, 42.28, 48.32, 54.36, 60.4, 66.4]
for r_val in r_values:
    plt.axvline(x=r_val, linestyle='--', color='green')

# Set labels and title
plt.xlabel('r')
plt.ylabel('z')
plt.title('2D Plot of Events in sr1 (-147.6515<z<-146.6515)')

plt.show()
```

```python
# Electrode     z position
# ------------------------
# Top Screen       +36 mm
# Anode             +8 mm
# Gate               0 mm
# Cathode        1486 mm
# Bottom Screen  1541 mm

# (from insturment paper https://arxiv.org/pdf/2402.10446.pdf)
```

```python
# Cut 'r' into bins and count the number of events in each bin
filtered_df['r_bin'] = pd.cut(filtered_df['r'], bins=r_values, include_lowest=True)
bin_counts = filtered_df['r_bin'].value_counts().sort_index()

# Calculate the percentage of events in each bin relative to the total number of events
total_events = bin_counts.sum()
bin_percentages = bin_counts / total_events * 100

# Create the 2D histogram plot
plt.bar(range(len(bin_percentages)), bin_percentages, width=1.0, tick_label='', align='edge', edgecolor='black')
plt.xticks(np.arange(len(r_values)), labels=r_values, rotation=45, ha='right')
plt.xlabel('r')
plt.ylabel('Percentage of Events')
plt.title('Percentage of Events in Each r Bin for sr1')
plt.show()
```
